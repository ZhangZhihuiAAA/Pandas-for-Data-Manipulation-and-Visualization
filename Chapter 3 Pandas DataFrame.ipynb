## Adding and Removing Columns in a DataFrame

```python
df = pd.concat([df, new_columns], axis=1)
df
```

```python
df.drop('price', axis=1, inplace=True)
df
```

```python
df1 = df.drop(0, axis=0, inplace=False)
df1
```

```python
df.drop(['season', 'taste'], axis=1, inplace=True)
df
```

## Reordering Columns in a DataFrame

```python
data = {
    'fruit': ['apple', 'banana', 'cherry'],
    'color': ['red', 'yellow', 'red'],
    'weight': [120, 150, 45]
}

df = pd.DataFrame(data)
df
```

```python
# Reordering columns using indexing
df = df[['color', 'weight', 'fruit']]
df
```

```python
data = {
    'fruit': ['apple', 'banana', 'cherry'],
    'color': ['red', 'yellow', 'red'],
    'weight': [120, 150, 45]
}

df = pd.DataFrame(data)
df
```

```python
# Reordering columns using reindex()
new_order = ['color', 'fruit', 'weight']
df = df.reindex(columns=new_order)
df
```

```python
# Reordering columns using the loc[] indexer
df = df.loc[:, ['weight', 'color', 'fruit']]
df
```

```python
# Reordering columns using the iloc[] indexer
df = df.iloc[:, [1, 0, 2]]
df
```

## Data Filtering Using Boolean Expressions

### Boolean Indexing

```python
data = {
    'Name': ['John', 'Alice', 'Bob', 'Emily', 'Michael'],
    'Age': [30, 25, 35, 27, 32],
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance'],
    'Salary': [50000, 60000, 55000, 45000, 70000]
}

df = pd.DataFrame(data)
df
```

```python
filtered_df = df[df['Age'] > 30]
filtered_df
```

```python
filtered_df = df[(df['Age'] > 30) & (df['Department'] == 'IT')]
filtered_df
```

```python
filtered_df = df[df['Department'].str.startswith('H')]
filtered_df
```

```python
filtered_df = df[df['Department'].isin(['HR', 'Finance'])]
filtered_df
```

```python
filtered_df = df[df['Salary'].isnull()]
filtered_df
```

```python
filtered_df = df[df['Salary'].notnull()]
filtered_df
```

## Arithmetic Operations on DataFrames

```python
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
}

df = pd.DataFrame(data)
df
```

```python
df_add_scalar = df + 10
df_add_scalar
```

```python
df_add_df = df + df
df_add_df
```

```python
df_multiply_df = df * df
df_multiply_df
```
