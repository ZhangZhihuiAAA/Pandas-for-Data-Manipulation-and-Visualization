```python
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
df = pd.DataFrame(data, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
```

```python
df['MEDV'] = target
```

```python
df
```

## Filtering by Logical Operators

```python
filtered_by_chas = df[df['CHAS'] == 1]
filtered_by_chas.head()
```

```python
filtered_greater_than = df[df['RM'] > 6]
filtered_greater_than.head()
```

```python
filtered_less_than = df[df['TAX'] < 300]
filtered_less_than.head()
```

```python
filtered_greater_equal = df[df['NOX'] >= 0.5]
filtered_greater_equal.head()
```

```python
filtered_and = df[(df['RM'] > 6) & (df['DIS'] <= 2.0)]
filtered_and.head()
```

```python
filtered_or = df[(df['RM'] > 6) | (df['DIS'] <= 2.0)]
filtered_or.head()
```

```python
filtered_not = df[~(df['RM'] > 6)]
filtered_not.head()
```

## Filtering by Numeric Methods

```python
filtered_query = df.query('RM > 6 and MEDV > 25')
filtered_query.head()
```

```python
filtered_between = df[df['TAX'].between(200, 400)]
filtered_between.head()
```

```python
filtered_isin_age = df[df['AGE'].isin([65.2, 90.0])]
filtered_isin_age.head()
```

## Filtering by Where Condition

```python
filtered_where = df.where(df['RM'] > 7)
filtered_where.head()
```

Remember that the where() method retains the original shape of the DataFrame and
replaces non-matching rows with NaN values. If you want to remove the rows where
the condition is not met, you can use the .dropna() method after applying .where():

```python
filtered_where_notnull = filtered_where.dropna()
filtered_where_notnull.head()
```

## Filtering by Date and Time Methods

```python
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
values = range(len(date_rng))
df = pd.DataFrame({'value': values}, index=date_rng)
```

```python
df
```

```python
filtered_date_index = df['2023-01-03':'2023-01-07']
filtered_date_index
```

By using the syntax df[‘2023-01-03’:’2023-01-07’], we are extracting
a subset of the original DataFrame that includes rows corresponding to the dates
from January 3rd, 2023, to January 7th, 2023, **inclusive**.

```python
filtered_loc = df.loc[df.index >= '2023-01-05']
filtered_loc
```

```python
filtered_query = df.query('index >= "2023-01-06"')
filtered_query
```

```python
date_rng = pd.date_range(start='2023-01-01 08:00', periods=10, freq='H')
values = range(len(date_rng))
df = pd.DataFrame({'value': values}, index=date_rng)
```

```python
date_rng = pd.date_range(start='2023-01-01 08:00', periods=10, freq='h')
values = range(len(date_rng))
df = pd.DataFrame({'value': values}, index=date_rng)
```

```python
filtered_between_time = df.between_time('10:00', '16:00')
filtered_between_time
```

## Filtering Null or Missing Values

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math': [85, np.nan, 72, 90, 65],
    'English': [90, 78, np.nan, 85, 92]
}

df = pd.DataFrame(data)
df
```

```python
missing_scores_mask = df.isna()
missing_scores_mask
```

```python
missing_scores_filtered = df[missing_scores_mask.any(axis=1)]
missing_scores_filtered
```

```python
non_missing_scores_mask = df.notna()
non_missing_scores_filtered = df[non_missing_scores_mask.all(axis=1)]
non_missing_scores_filtered
```
